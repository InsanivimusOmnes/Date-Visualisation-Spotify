```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('Rank.xlsx')
ds = pd.read_excel('top10.xlsx')
```

```python
df.describe
```

```python
sns.histplot(df["Year published"], binwidth=1)
'''On the graph you can see how fast the music industry is developing. There are more and more creative people with good hearing and voice'''
```

```python
sns.barplot(data=df, x="Year published", y="Rank")
'''It can be assumed that people are not afraid of something new, but always remember the good old classics.'''
```

```python
sns.barplot(x="Album", y="Streams", data=df)
'''According to the number of views, the most popular albums were hip hop trends. Prevailing in the West'''
```

```python
pivot = ds.pivot_table(
    index=['Rank', 'Artist'],
    columns=["Year published"],
    values=["Streams"],
    aggfunc=np.average)
sns.heatmap(pivot)
'''It can be assumed that the period of time 2016 - 2019 was more fruitful for the world of music and new talents were revealed'''
```
